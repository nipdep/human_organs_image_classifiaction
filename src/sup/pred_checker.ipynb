```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array, load_img
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
def pred_sample(model_path, model_weight_path, img_path):
    clToInt_dict={0: 'Brain', 1: 'Eye', 2: 'Heart', 3: 'Kidney', 4: 'Skeleton'}
    model = load_model(model_path)
    model.load_weights(model_weight_path)

    x_img = load_img(img_path, target_size=(224, 224))
    x = img_to_array(x_img,dtype='float32')/225.
    x = np.expand_dims(x, axis=0)
    result = model.predict(x)
    img_class = np.argmax(result[0])
    str_img_class = clToInt_dict[img_class]

    plt.imshow(x_img)
    plt.title(str_img_class)
    plt.show()

    return
```

```python
model_path = '../../h5_files/models/20200822234813mobileNet_v2_transfer_learn_2.2.h5'
model_weight_path = '../../h5_files/weights/20200822234813mobileNet_v2_transfer_learn_2.2.h5'
img_path = '../../sample_images/skull-2.jpg'
```

```python
pred_sample(
    model_path,
    model_weight_path,
    img_path
)
```
